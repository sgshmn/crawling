# G마켓 상품 정보 수집
 
- 상품명, 가격

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By

import time
import pandas as pd
```

```python
gm_url = 'http://corners.gmarket.co.kr/Bestsellers'

driver = wb.Chrome()
driver.get(gm_url)
```

```python
driver.find_element(By.CLASS_NAME, value='itemname').text
```

```python
itemNms = driver.find_elements(By.CLASS_NAME, value='itemname')
itemPrices = driver.find_elements(By.CSS_SELECTOR, value='.s-price>strong>span>span')

for i in range(len(itemNms)):
    print(itemNms[i].text)
    print(itemPrices[i].text)
```

```python
for i in range(13):
    # 카테고리를 누를 때 마다 페이지 내용이 바뀌기 때문에
    # 전에 수집한 카테고리 정보를 사용할 수 없다
    # 매번 페이지가 바뀔 때마다 카테고리를 수집할 수 있게
    # 반복문 안으로 넣어주면 카테고리를 하나씩 클릭할 수 있다
    tabs = driver.find_elements(By.CSS_SELECTOR, value='#categoryTabG > li')
    
    tabs[i].click()
    time.sleep(1)
```

### 패션의류 ~ e쿠폰까지 상품명, 가격 출력

```python
#  각 리스트에 데이터를 저장한 후 데이터 프레임 만들어보기
categorylist = []
itemNmlist = []
itemPricelist = []

for i in range(1, 13):
    tabs = driver.find_elements(By.CSS_SELECTOR, value='#categoryTabG > li')
    tabs[i].click()
    
    # 카테고리를 클릭하면 다시 자료를 수집해야 한다
    tabs = driver.find_elements(By.CSS_SELECTOR, value='#categoryTabG > li')
    # print(tabs[i].text)
    # print()
    
    itemNms = driver.find_elements(By.CLASS_NAME, value='itemname')
    itemPrices = driver.find_elements(By.CSS_SELECTOR, value='.s-price>strong>span>span')

    for j in range(len(itemNms)):
        categorylist.append(tabs[i].text)
        itemNmlist.append(itemNms[j].text)
        itemPricelist.append(itemPrices[j].text)
        
        # print(itemNms[j].text)
        # print(itemPrices[j].text)
    
    time.sleep(1)
    
gm_item_dic = {
    '카테고리':categorylist,
    '상품명':itemNmlist,
    '가격':itemPricelist
}

gm_item_df = pd.DataFrame(gm_item_dic)

gm_item_df
```

# 코드통합

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from urllib.request import urlretrieve

import time
import pandas as pd
import os

# 이미지를 수집할 브라우저 띄우기
keyword = '오로라'
nv_url = f'https://search.naver.com/search.naver?sm=tab_sug.top&where=image&query={keyword}'

driver = wb.Chrome()
driver.get(nv_url)

# 브라우저 로딩 시간 기다리기
time.sleep(3)

# 페이지 스크롤 5번 내리기
for i in range(5):
    driver.find_element(By.TAG_NAME, value='body').send_keys(Keys.END)
    time.sleep(2)

# 이미지 주소 수집
imgs2 = driver.find_elements(By.CSS_SELECTOR, value='._image._listImage')

# 폴더 생성
if not os.path.isdir(keyword):
    print('폴더생성')
    os.mkdir(keyword)

# 주소를 srclist에 담기
srclist = []
for img in imgs2:
    if img.get_attribute('src')[0] != 'd':
        srclist.append(img.get_attribute('src'))

# srclist의 이미지 주소를 폴더 안에 저장
for i in range(len (srclist)):
    fileNm = f'./{keyword}/{keyword}{i+1}.jpg'
    
    urlretrieve(srclist[i], fileNm)
    
    time.sleep(1)

# 브라우저 종료
driver.close()
```
