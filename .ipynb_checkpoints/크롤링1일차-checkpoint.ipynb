# BeautifulSoup 라이브러리
- 응답받은 HTML문서 안의 데이터를 추출할 수 있는 라이브러리
1. 문자열 HTML문서 -> 파이썬객체로 변환
2. HTML객체를 반환하는 함수를 이용하여 데이터 수집

```python
from bs4 import BeautifulSoup as bs
```

```python
nv_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%83%80%EC%BD%94%EC%95%BC%EB%81%BC'

res = req.get(nv_url)
```

```python
# bs(변환할 데이터, 변환방식) 변환방식은 3가지가 있다
# bs : 문자열 HTML문서를 파이썬객체로 변환
# 변환방식 : lxml > html.parser > html5lib
html = bs(res.text, 'lxml') 
```

```python
print(type(res.text))
print(type(html))
```

```python
# 태그선택자(태그이름), 아이디선택자(#), 클래스선택자(.)
# 자식선택자(>), 자손선택자(빈칸), 형제선택자(~), 인접형제선택자(+)

html.select_one('title').text
```

```python
# 클래스 선택자를 이용하여 데이터 수집
# 양쪽 끝 공백제거 : strip()

html.select_one('a.lnk_head').text.strip()
# select_one(CSS선택자) : 선택자와 같은 HTML객체 중 처음 발견한 하나를 반환
```

```python
adList = html.select('a.lnk_head')
# select(CSS선택자) : 선택자와 같은 모든 HTML객체를 리스트타입으로 반환한다

adList[0].text
```

```python
for i in range(15):
    print(adList[i].text.strip())
```

```python
for i in adList:
    print(i.text.strip())
```

# 과제 : 네이버 뉴스 수집
- 뉴스 제목, 뉴스 내용을 수집하고 출력



```python
import requests as req
from bs4 import BeautifulSoup as bs

# 웹페이지의 HTML문서를 요청/응답
news_url = 'https://n.news.naver.com/mnews/article/031/0000686189?sid=101'

## 요청하는 브라우저의 정보를 설정
header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}
res = req.get(news_url, headers=header)

# 문자열 HTML문서를 파이썬 객체로 변환
anjwl = bs(res.text, 'lxml')

# HTML에 저장한 정보 중에 필요한 정보를 수집
# select, select_one함수 활용
title = anjwl.select_one('h2.media_end_head_headline').text # 태그를 찾을 때는 우클릭 검사 기능을 이용하자
scripts = anjwl.select_one('#dic_area').text.strip()


print('제목:',title, end="\n\n")
print('내용:',scripts)





      
```
