# 멜론 음원차트 수집

- 순위(rank), 곡명(songNm), 가수(singer), 이미지경로(src)
1. 4가지 정보 수집
2. 수집한 정보를 리스트에 저장(텍스트데이터)
3. 4개의 리스트를 하나의 딕셔너리로 생성
4. 딕셔너리를 데이터프레임(표?)으로 생성

```python
import requests as req
import pandas as pd # 데이터프레임을 만들 때 사용

from bs4 import BeautifulSoup as bs
```

```python
melon_url = 'https://www.melon.com/chart/'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = req.get(melon_url, headers=header)

html = bs(res.text, 'lxml') # 문자열 html문서와 바꿀 방법
```

```python
ranks = html.select('td>div>span.rank')
songNms = html.select('div.ellipsis.rank01 a')
singers = html.select('div.ellipsis.rank02>span')
imgs = html.select('a.image_typeAll>img')

len(ranks), len(songNms), len(singers), len(imgs)
```

```python
# HTML태그의 속성값 접근
imgs[0].get('src')
```

```python
ranklist = []
songNmlist = []
singerlist = []
srclist = []

for i in range(len(ranks)):
#    print(ranks[i].text, songNms[i].text, singers[i].text, imgs[i].get('src'))
    # 각 리스트에 데이터를 저장하기
    ranklist.append(ranks[i].text)
    songNmlist.append(songNms[i].text)
    singerlist.append(singers[i].text)
    srclist.append(imgs[i].get('src'))
```

```python
# 딕셔너리 : key와 value 형태의 자료형
# 구조 : {key : value, key : value, ...}

dic = {"name" : "MJ"}
dic["name"]
```

```python
# 딕셔너리를 데이터프레임으로 만들려고 한다면 딕셔너리를 생성할 때 각 리스트의 개수는 같아야한다!

melon_dic = {
    "순위" : ranklist,
    "곡명" : songNmlist,
    "아티스트" : singerlist,
    "이미지주소" : srclist
}

melon_dic
```

```python
# 딕셔너리를 활용한 데이터프레임 생성
# pd.DataFrame(딕셔너리변수)
melon_df = pd.DataFrame(melon_dic)

melon_df
```

```python
# CSV파일로 저장
# melon_df.to_csv(저장할 경로, 인코딩, ...)
# 인코딩 : 컴퓨터가 이해할 수 있는 언어로 변경하는 속성(euc-kr, utf-8, utf-8-sig, ...)
# euc-kr : 한글, 영어, 숫자 kr, us 등등 있다
# utf-8 : 전세계 모든 언어
melon_df.to_csv('melon_top100.csv', encoding='euc-kr', index=False)
```

# 네이버 영화 리뷰 수집

1. 하나의 페이지에 있는 리뷰 수집
2. 1~5 페이지에 있는 리뷰 수집
3. 모든 페이지에 있는 리뷰 수집

```python
import requests as req
import pandas as pd
import time # 딜레이를 사용할 때

from bs4 import BeautifulSoup as bs
```

```python
review_url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = req.get(review_url, headers=header)

html = bs(res.text, 'lxml')
```

```python
reviews = html.select('div.score_reple>p>span')

for i in reviews:
    # 조건문을 활용해서 '관람객'을 제외한 데이터만 출력
    # '관람객' in i.text : 리뷰에 '관람객'이 있으면 제외된다
    if i.text != '관람객':
        print(i.text.strip())
    
```

```python
# 1~5 페이지 HTML문서 요청

for pageNo in range(1,6):
    review_url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false'

    header = {
        'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
    }

    res = req.get(review_url, headers=header,params={'page':pageNo})

    html = bs(res.text, 'lxml')
    
    print('='*10, f'[페이지번호:{pageNo}]', '='*10)
    
    reviews = html.select('div.score_reple>p>span')

    for i in reviews:
        if i.text != '관람객':
            print(i.text.strip())
    
    
    # 1초간 딜레이
    time.sleep(1)
    
    
```

### 전체 페이지 수 구하기
1. 총 리뷰수 수집 --> '30,186'
2. 리뷰수 데이터의 ','를 ''로 변환 --> '30186'
3. 정수형 데이터로 변환 --> 30186
4. 10으로 나누기 -> 페이지 수 --> 3018
* 나머지가 0보다 클 경우 페이지수 + 1

```python
review_url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = req.get(review_url, headers=header)

html = bs(res.text, 'lxml')

rvCnt = html.select_one('.total>em').text

rvCnt = int(rvCnt.replace(',', ''))
# replace()는 변수에 영향을 안 준다
# 대체한 값을 유지하려면 변수에 저장해야한다

rvPs = (rvCnt-1)//10 + 1

print(rvPs)
```

# 코드 통합

```python
import requests as req
import pandas as pd
import time # 딜레이를 사용할 때

from bs4 import BeautifulSoup as bs
from tqdm.notebook import tqdm as tn

review_url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = req.get(review_url, headers=header)

html = bs(res.text, 'lxml')

rvCnt = html.select_one('.total>em').text

rvCnt = int(rvCnt.replace(',', ''))

rvPs = (rvCnt-1)//10 + 1

# 파일 객체 생성
f = open('review2.txt', 'w')

# 모든 페이지 요청
for pageNo in tn(range(1,rvPs+1)):
    res = req.get(review_url, headers=header,params={'page':pageNo})

    html = bs(res.text, 'lxml')
    
    print('='*10, f'[페이지번호:{pageNo}]', '='*10)
    
    reviews = html.select('div.score_reple>p>span')

    for i in reviews:
        if i.text != '관람객':
            print(i.text.strip())
            
            # 리뷰 데이터 저장
            f.write(i.text.strip())
    
    time.sleep(1)

# 파일 객체 닫기
f.close()
```
