# 영화 평점 수집

1. 영화제목과 평점을 수집하고 데이터프레임으로 만들기
2. 날짜별로 영화제목과 평점을 수집하고 데이터프레임으로 만들기

```python
import requests as req
import pandas as pd
import time # 딜레이를 사용할 때

from bs4 import BeautifulSoup as bs
```

```python
rate_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220719'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = req.get(rate_url, headers=header)

html = bs(res.text, 'lxml')
```

```python
# 영화제목, 평점 수집하고 출력하기

titles = html.select('.tit5>a')
rates = html.select('td.point')

titlelist = []
ratelist = []




for i in range(len(titles)) :
#    print(titles[i].text, rates[i].text)
    titlelist.append(titles[i].text)
    ratelist.append(rates[i].text)
```

```python
mv_rate_dic = {
    '영화명':titlelist,
    '평점':ratelist
}

mv_rate_df = pd.DataFrame(mv_rate_dic)

mv_rate_df
```

```python
# 날짜 생성하는 함수
# pd.date_range(start='2022-06-01', end='2022-06-30')
dates = pd.date_range(start='2022-06-01', periods=30)

datelist = []

for i in range(len(dates)) :
    datelist.append(dates[i].strftime('%Y%m%d'))
```

```python
titlelist = []
ratelist = []
daylist = []

for date in datelist :
    rate_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur'

    header = {
        'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
    }

    res = req.get(rate_url, headers=header, params={'date':date})

    html = bs(res.text, 'lxml')
    
    titles = html.select('.tit5>a')
    rates = html.select('td.point')

    for i in range(len(titles)) :
        titlelist.append(titles[i].text)
        ratelist.append(rates[i].text)
        daylist.append(date)
        
    time.sleep(1)

mv_rate_dic = {
    '영화명':titlelist,
    '평점':ratelist,
    '날짜':daylist
}

mv_rate_df = pd.DataFrame(mv_rate_dic)

```
