# 유튜브 영상 수집

- 영상url, 영상 제목, 조회수 데이터 수집
- 데이터 프레임으로 만들기

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd

```

```python
yt_url = 'https://www.youtube.com/c/pinokimusic/videos'

driver = wb.Chrome()
driver.get(yt_url)
```

### 실습 - 영상주소, 제목, 조회수 출력

```python
# shorts 영상은 빼기
videos = driver.find_elements(By.CSS_SELECTOR, value='a#video-title')

srclist = []
titlelist = []
viewCntlist = []

for video in videos:
    if video.get_attribute('aria-label')[-1] != '회':
        continue
    srclist.append(video.get_attribute('href')) # 주소
    titlelist.append(video.text) # 제목
    inform = video.get_attribute('aria-label') 
    j = 0
    for i in inform[::-1]:
        if i == " ":
            break
        j += 1
    
    viewCntlist.append(inform[-j:-1].replace(",", "")) # 조회수
    time.sleep(1)

```

```python
# shorts 영상 포함하기
videos = driver.find_elements(By.CSS_SELECTOR, value='a#video-title')

srclist = []
titlelist = []
viewCntlist = []

for video in videos:
    
    srclist.append(video.get_attribute('href')) # 주소
    titlelist.append(video.text) # 제목
    inform = video.get_attribute('aria-label') 
    si = inform.rfind("조회수") + 4
    ei = inform.rfind("회") # rfind 오른쪽부터 찾기
    viewCntlist.append(inform[si:ei]) # 조회수
    time.sleep(1)

```

```python
# shorts 영상 포함하기 2
videos = driver.find_elements(By.CSS_SELECTOR, value='a#video-title')

hreflist = []
titlelist = []
viewCntlist = []

for video in videos:
    
    hreflist.append(video.get_attribute('href')) # 주소
    titlelist.append(video.text) # 제목
    inform = video.get_attribute('aria-label') 
    j = 0
    for i in inform[::-1]:
        j += 1
        if i == "회":
            break
    k = 0
    for i in inform[-j:0:-1]:
        if i == " ":
            break
        k += 1
    
    viewCntlist.append(inform[-j-k:-j].replace(",", "")) # 조회수
    time.sleep(1)
```

```python
videos_dic = {
    "제목" : titlelist,
    "조회수" : viewCntlist,
    "영상주소" : hreflist
        
}

videos_df = pd.DataFrame(videos_dic)
videos_df
```

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
map_url = 'https://map.naver.com/v5/?c=14140822.8653826,4508397.7630627,15,0,0,0,dh'
driver = wb.Chrome()
driver.get(map_url)
time.sleep(2)
# 지도 검색어 입력하기
driver.find_element(By.CSS_SELECTOR, value='input.input_search').send_keys('중국집\n')
time.sleep(1)
# 검색결과 창 전환
driver.switch_to.frame('searchIframe')
# 검색결과 중 가게하나를 클릭
stores = driver.find_elements(By.CSS_SELECTOR, value='span.place_bluelink.OXiLu')
time.sleep(1)
for store in stores:
    store.click()
    time.sleep(1)
    # 이전페이지 창 전환
    driver.switch_to.default_content()
    # 상세페이지 창 전환
    driver.switch_to.frame('entryIframe')
    try:
        # 상세페이지에 있는 가게정보 수집
        storeNm = driver.find_element(By.CSS_SELECTOR, value='span._3XamX').text
    except:
        storeNm = ''
    print('가게명:', storeNm)
    # 이전페이지 창 전환
    driver.switch_to.default_content()
    # 검색결과 창 전환
    driver.switch_to.frame('searchIframe')
    time.sleep(1)
driver.close()
```
