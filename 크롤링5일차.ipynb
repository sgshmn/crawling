```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd

```

```python
yt_url = 'https://www.youtube.com/c/pinokimusic/videos'

driver = wb.Chrome()
driver.get(yt_url)
```

```python
# shorts ì˜ìƒì€ ë¹¼ê¸°
videos = driver.find_elements(By.CSS_SELECTOR, value='a#video-title')

srclist = []
titlelist = []
viewCntlist = []

for video in videos:
    if video.get_attribute('aria-label')[-1] != 'íšŒ':
        continue
    srclist.append(video.get_attribute('href')) # ì£¼ì†Œ
    titlelist.append(video.text) # ì œëª©
    inform = video.get_attribute('aria-label') 
    j = 0
    for i in inform[::-1]:
        if i == " ":
            break
        j += 1
    
    viewCntlist.append(inform[-j:-1].replace(",", "")) # ì¡°íšŒìˆ˜
    time.sleep(1)

```
